### 1. LeRobot 및 Hugging Face 관련 패키지 설치
LeRobot 데이터셋 편집 및 학습에 필요한 `lerobot` 패키지와 Hugging Face Hub 연동을 위한 `huggingface_hub` 패키지를 설치



```python
# !pip install -U "lerobot"
# !pip install -U huggingface_hub

!pip install -U "lerobot"
!pip install -U huggingface_hub

import pathlib
import lerobot

p = pathlib.Path(lerobot.__file__).parent / "policies/groot/groot_n1.py"
text = p.read_text()

repls = {
    'backbone_cfg: dict = field(init=False, metadata={"help": "Backbone configuration."})':
    'backbone_cfg: dict = field(default=None, init=False, metadata={"help": "Backbone configuration."})',

    'action_head_cfg: dict = field(init=False, metadata={"help": "Action head configuration."})':
    'action_head_cfg: dict = field(default=None, init=False, metadata={"help": "Action head configuration."})',

    'action_horizon: int = field(init=False, metadata={"help": "Action horizon."})':
    'action_horizon: int = field(default=None, init=False, metadata={"help": "Action horizon."})',

    'action_dim: int = field(init=False, metadata={"help": "Action dimension."})':
    'action_dim: int = field(default=None, init=False, metadata={"help": "Action dimension."})',
}

for old, new in repls.items():
    if old not in text:
        print("이미 패치됐거나 코드가 다릅니다:", old)
    text = text.replace(old, new)

p.write_text(text)
print("GROOT dataclass patch done:", p)
```
